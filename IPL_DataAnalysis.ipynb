# TOP 5 MOM PLAYERS 

```python
temp_data=match['player_of_match'].value_counts().head()
print(temp_data)
#sns.barplot(x=data['player_of_match'].value_counts().head().index,y=data['player_of_match'].value_counts().head().values,data=data)
sns.barplot(x=temp_data.index,y=temp_data.values,data=match)

# Add data labels to each bar
for i, value in enumerate(temp_data.values):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Plot initiation and labeling 
plt.title("Top 5 MoM Players")
plt.xticks(rotation=90)
plt.xlabel("No of Matches")
plt.ylabel("Player Name")
# Display the plot 
plt.show()
```

# Top 10 Batsmen by Runs Scored

```python
import matplotlib.pyplot as plt

top_batsman = delivery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values('batsman_runs', ascending=False).head(10)
top_batsman.set_index('batsman', inplace=True)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Create the bar chart
ax1.bar(top_batsman.index, top_batsman['batsman_runs'], color='green')
ax1.set_ylabel('Runs')
ax1.set_xlabel('Batsman')
ax1.set_title('Top 10 Batsmen by Runs Scored')

ax2 = ax1.twinx()

# Create the line chart
ax2.plot(top_batsman.index, top_batsman['batsman_runs'], color='red', marker='o')
#ax2.set_ylabel('Runs')
ax2.grid(False)

# Adjust the y-axis limits for the line chart
min_runs = top_batsman['batsman_runs'].min()
max_runs = top_batsman['batsman_runs'].max()
ax2.set_ylim(min_runs - 100, max_runs + 100)

# Add data labels to the line chart
for i, runs in enumerate(top_batsman['batsman_runs']):
    ax2.annotate(runs, (i, runs), ha='center', va='bottom')

plt.xticks(rotation=45)
# Display the plot 
plt.show()
```

# MOST SIXES 

```python
temp_df = delivery.groupby('batsman')['batsman_runs'].agg(lambda x:(x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).head(10).reset_index(drop=True)
temp_df
sns.barplot(x=temp_df['batsman'],y=temp_df['batsman_runs'],data=temp_df)
# plotting initiation 
plt.title("Most sixes by a Player")
plt.xticks(rotation=90)
plt.xlabel("Batsman")
plt.ylabel("Number of 6's")
plt.show()
```

# MOST WICKETS TAKEN BY BOWLER 

```python
temp_df = delivery.groupby('bowler')['is_wicket'].agg('sum').reset_index().sort_values(by='is_wicket', ascending=False).reset_index(drop=True).head(10)
sns.barplot(x=temp_df['bowler'],y=temp_df['is_wicket'],data=temp_df)
plt.title("Most wickets by a bowler")
plt.xticks(rotation=90)
plt.xlabel("Bowler")
plt.ylabel("Number of wickets")
plt.show()
```

# MOST FOURS BY A PLAYER 

```python
temp_four = delivery.groupby('batsman')['batsman_runs'].agg(lambda x:(x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).head(10).reset_index(drop=True)
sns.barplot(x=temp_four['batsman'],y=temp_four['batsman_runs'],data=temp_four)
# plotting initiation 
plt.title("Most fours  by a Player")
plt.xticks(rotation=90)
plt.xlabel("Batsman")
plt.ylabel("Number of 4's")
plt.show()
```
